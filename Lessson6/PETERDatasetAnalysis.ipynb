#**Libraries**

```python
import requests
import pandas as pd
```

#**Data Collection**

```python
df = pd.read_csv('https://data.cityofnewyork.us/resource/yjxr-fw8i.csv?year=2018/19&boro=1&$offset=0')
for y in range(1000,100000,1000):
  link = f"https://data.cityofnewyork.us/resource/yjxr-fw8i.csv?year=2018/19&boro=1&$offset={y}"
  print(link)
  d = pd.read_csv(link)
  df = pd.concat([df, d])
```
