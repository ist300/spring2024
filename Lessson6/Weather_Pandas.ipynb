````python
from IPython.display import display, HTML, Image
from ipywidgets import interact_manual
from datetime import datetime
import pandas as pd
import requests

def geocode(location):
    query_string = {'q' : location, 'format': 'json'}
    url='https://nominatim.openstreetmap.org/search' 
    response = requests.get(url, params = query_string)
    response.raise_for_status()
    geodata = response.json()
    return geodata

def getweather(weather_key, lat, lon):
    query_string = {'lat': lat, 'lon': lon, 'appid': weather_key, 'units': 'imperial'}
    url = 'https://api.openweathermap.org/data/2.5/onecall'
    response = requests.get(url, params = query_string)
    weather = response.json()
    return weather


def transform(df):
    new = []
    for row in df.to_records():
        day = datetime.fromtimestamp(row['dt']).strftime("%A")
        new.append( {'Day' : day, 'Low' : int(row['temp.min']), 'High': int(row['temp.max'])})
    return pd.DataFrame(new)


weather_key = '67f9d45c32dd086df27f2b7a90370a8e'
display(HTML("<h1>Get 7-Day Forecast</h1>"))

@interact_manual(location = " ")
def main (location):
    
    iconurl = 'http://openweathermap.org/img/wn/{icon}@2x.png'
    geo = geocode(location)
    lat, lon = geo[0]['lat'],geo[0]['lon']
    weather = getweather(weather_key, lat, lon)
    icon = weather['current']['weather'][0]['icon']
    desc = weather ['current']['weather'][0]['description']
    temp = int(weather['current']['temp'])
    
    display(HTML(f"<h2>Current Weather For {location}</h2>"))
    display(HTML(f"<p>{desc} with a temperature of {temp} deg F.".title()))
    display(Image(iconurl = iconurl))
    display(HTML(f"<h2> 7-Days Forecast for {location}</h2>"))
    df = pd.json_normalize(weather['daily'])
    df2 = transform(df)
    display(df2)
````
